```python
import pandas as pd
import numpy as np
import nltk

from nltk import word_tokenize
```

```python
nltk.download("punkt")
```

```python
df = pd.read_csv('/content/bbc-text.csv')
```

```python
df
```

```python
# convert documents into sequences of ints/ids/indices

idx = 0
word2idx = {}
tokenized_docs= []
for doc in df['text']:
  words = word_tokenize(doc.lower())
  doc_as_int = []
  for word in words:
    if word not in word2idx:
      word2idx[word] = idx
      idx = idx+1

    doc_as_int.append(word2idx[word])
  tokenized_docs.append(doc_as_int)

```

```python
# Reverse mapping

idx2word = {v:k for k, v in word2idx.items()}
```

```python
# number of documnets

N = len(df['text'])
```

```python
# Number of words

V = len(word2idx)
```

```python
# instantiate term frequency matrix

tf = np.zeros((N,V))
```

```python
# Populate term frequency matrix

for i, doc_as_int in enumerate(tokenized_docs):
  for j in doc_as_int:
    tf[i, j] +=1
```

```python
# Compute IDF

document_freq = np.sum(tf >0, axis = 0)
idf = np.log(N/ document_freq)
```

```python
# Compute TF-IDF

tf_idf = tf * idf
```

```python
np.random.seed(42)
```

```python
# pick a random document, show the top 5 terms

i = np.random.choice(N)
row = df.iloc[0]
print("label:", row['category'])
print("Text:", row['text'].split("\n",1)[0])
print("Top 5 terms")

scores = tf_idf[i]
indices = (-scores).argsort()

for j in indices[:5]:
  print(idx2word[j])
```
