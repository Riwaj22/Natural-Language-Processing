```python
text = """Once upon a time in a quaint village nestled between rolling hills and lush forests, there lived a young girl named Ella. Ella was known for her adventurous spirit and curiosity about the world beyond her village. She would often spend hours exploring the woods, discovering hidden trails and secret meadows where wildflowers bloomed in abundance.
One sunny morning, while wandering deeper into the forest than usual, Ella stumbled upon an ancient stone path overgrown with moss and ferns. Intrigued, she followed the path which led her to a secluded glade with a shimmering pond at its center. The water sparkled like liquid diamonds under the golden sunlight filtering through the trees."""
```

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(text)
```

```python
tokens = [token.text for token in doc]
print(tokens)
```

Text Cleaning

```python
import string
from spacy.lang.en.stop_words import STOP_WORDS

# Initialize an empty dictionary to store word frequencies
word_freq = {}

# Get the set of stop words and punctuation
stop_words = list(STOP_WORDS)
punctuation = string.punctuation

# Assuming 'doc' is your processed spaCy document
for word in doc:
    # Check if the word is not a stop word and not punctuation
    if word.text.lower() not in stop_words and word.text.lower() not in punctuation:
        # Update word frequency dictionary
        if word.text not in word_freq:
            word_freq[word.text] = 1
        else:
            word_freq[word.text] += 1

# Now 'word_freq' dictionary contains word frequencies (excluding stop words and punctuation)
print(word_freq)
```

```python
max_freq = max(word_freq.values())
```

```python
print(max_freq)
```

```python
for word in word_freq.keys():
  word_freq[word] = word_freq[word] / max_freq
```

```python
print(word_freq)
```

**Sentence Tokenization**

```python
sent_tokens = [sent for sent in doc.sents]
print(sent_tokens)
```

```python
sent_score = {}

for sent in sent_tokens:
    total_score = 0  # Initialize total score for the current sentence
    for word in sent:
        if word.text.lower() in word_freq.keys():  # Check if the word is in word_freq dictionary
            total_score += word_freq[word.text.lower()]  # Add word frequency to total score

    # Store the total score of the sentence in sent_score dictionary
    sent_score[sent.text] = total_score

# Now 'sent_score' dictionary contains scores for each sentence based on word frequencies
print(sent_score)
```

Select 30% sentences with maximum scores

```python
from heapq import nlargest
```

```python
len(sent_score) * 0.3
```

```python
summary = nlargest(n =2, iterable = sent_score, key = sent_score.get)
```

```python
print(summary)
```

```python
import spacy

# Load spaCy language model
nlp = spacy.load('en_core_web_sm')

# Assuming 'summary' is a list of words
summary_list = summary  # Replace with your list of words

# Convert list of words into a single string
summary_text = ' '.join(summary_list)

# Process the summary text to obtain a spaCy Doc object
summary_doc = nlp(summary_text)

# Extract words from the processed summary (assuming 'summary_doc' is a spaCy Doc object)
final_summary = [word.text for word in summary_doc]

# Print the final summary (list of words)
print(final_summary)
```
